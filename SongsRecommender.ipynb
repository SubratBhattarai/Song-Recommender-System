```python
## Importing Libraries

import pandas as pd
import numpy as np
import string
```

```python
## Reading CSV File and observing top 5 songs
df = pd.read_csv('Top 500 Songs.csv', encoding='unicode_escape')
df.head()
```

```python
## Converting a df titles to lowercase
def lowering(content):
        content = content.lower()
        return content

df['title'] = df['title'].apply(lowering)
```

```python
## Droping useless columns --> But could be used to make system more accurate
df = df.drop(['appears on', 'streak', 'position', 'released'], axis=1)
df.head()
```

```python
## Creating own index
df['index'] = range(1, len(df)+1)
```

```python
## Importing function to convert string to number
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
## Creating a combined content
combined_feature = df['title'] + ' ' + df['artist'] + ' ' + df ['writers'] + ' ' + df['producer']
```

```python
TFV = TfidfVectorizer()
```

```python
## Converting
converted = TFV.fit_transform(combined_feature)
```

```python
print(converted)
```

```python
## Difflab to get close name matches, and cosine similarity to find the similarity distance

from difflib import get_close_matches
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Converting titles to list
title = df['title'].tolist()
```

```python
## Inputing name and Lowercasing

name = input("Enter the name: ")
name = name.lower().strip()
print(name)
```

```python
# Getting similarity value
similarScore = cosine_similarity(converted)
```

```python
## Try-except to get index and value name
try:
    index = df[df['title'] == name]['index'].values[0]
    print(index)
except:
    print("Song doesn't exist in Database.")
```

```python
#Listing and enumarting the score
similarSongs = list(enumerate(similarScore[index]))
```

```python
## Arranging the scores in descending order
sortScore = sorted(similarSongs, reverse= True, key=lambda x:x[1])
```

```python
## Printing the song name using index
i = 1


for index, _ in sortScore[1:20]:
    try: 
        print(str(i) + '. ' + str((df[df['index'] == index]['title'].values[0])))
        i +=1
    except:
        print('\n')
```
